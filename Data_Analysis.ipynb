## Add a dummy for names of apps with frequently used words:

```python
datalast = pd.read_csv("data/estimation_data.csv", sep=";", index_col="Unnamed: 0")
```

```python
pd.DataFrame(datalast["app_name"]).apply(','.join, axis=1)
```

```python
datalast["words"]=datalast["app_name"].str.replace("[^A-Za-z ]", " ").str.replace("  ", " ").str.lower()
```

```python
datalast["words"]
```

```python
app_str = pd.DataFrame(datalast.loc[datalast["words"].isnull()==False, "words"]).drop_duplicates().apply(' '.join, axis=0)[0]
```

```python
freq_str = app_str.split(" ")
```

```python
freq_str= [x for x in freq_str if len(x)>0]
```

```python
len(freq_str)
```

```python
freq_str.value_counts()
```

```python
att =  [ {word: freq_str.count(word)} for word in set(freq_str) ]
```

```python
att_new = {k: v for d in att for k, v in d.items()}
```

```python
att_new
```

```python
datalast = datalast[datalast["words"].isnull()==False]
```

```python
datalast["app_name_freq"] = sum([att_new[word]/3276 for word in datalast["words"].str.split()])
```

```python
datalast["freq_dum"]  = datalast["words"].str.lower().str.split().apply(lambda x: sum([att_new[word]/740 for word in x]))
```
